```python
# Laboratorio 6 - Algoritmo A*
# Autor: [Tu nombre]
# Curso: Sistemas Inteligentes - Ingeniería de Sistemas

import heapq  # Para manejar la cola de prioridad

# ------------------------------
# 1. Grafo y heurística
# ------------------------------
# Grafo definido como diccionario: nodo -> [(vecino, costo)]
grafo = {
    'A': [('B', 1), ('C', 4)],
    'B': [('D', 2), ('E', 5)],
    'C': [('F', 3)],
    'D': [],
    'E': [('F', 1)],
    'F': []
}

# Heurística: estimación de costo desde cada nodo hasta el objetivo
heuristica = {
    'A': 6,
    'B': 4,
    'C': 2,
    'D': 3,
    'E': 1,
    'F': 0
}

# ------------------------------
# 2. Función A* (búsqueda informada)
# ------------------------------
def a_star(inicio, objetivo):
    # Cola de prioridad: (f(n), nodo)
    cola = [(0, inicio)]
    visitados = set()
    costos = {inicio: 0}      # g(n): costo acumulado desde inicio
    padres = {inicio: None}   # Para reconstruir el camino

    while cola:
        _, actual = heapq.heappop(cola)  # Nodo con menor f(n)
        if actual == objetivo:
            break
        visitados.add(actual)

        for vecino, costo in grafo[actual]:
            nuevo_costo = costos[actual] + costo  # g(n) acumulado
            if vecino not in costos or nuevo_costo < costos[vecino]:
                costos[vecino] = nuevo_costo
                prioridad = nuevo_costo + heuristica[vecino]  # f(n) = g(n)+h(n)
                heapq.heappush(cola, (prioridad, vecino))
                padres[vecino] = actual

    return padres

# ------------------------------
# 3. Ejecutar A* desde A hasta F
# ------------------------------
resultado = a_star('A', 'F')
print("Diccionario de padres:", resultado)

# Reconstruir camino óptimo
camino = []
nodo = 'F'
while nodo:
    camino.append(nodo)
    nodo = resultado[nodo]
camino.reverse()
print("Camino óptimo:", camino)

# ------------------------------
# 4. Ejemplo de análisis adicional
# Cambiar nodo inicial a B y modificar heurística
# ------------------------------
heuristica['C'] = 5  # Modificamos la heurística de C
resultado2 = a_star('B', 'F')
camino2 = []
nodo = 'F'
while nodo:
    camino2.append(nodo)
    nodo = resultado2[nodo]
camino2.reverse()
print("\nPrueba con nodo inicial 'B' y heurística modificada:")
print("Camino óptimo:", camino2)
```
